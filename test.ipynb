```python
def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 * (x - mu)**2 / sigma**2)
# Use NumPy again for visualization
x = np.arange(-7, 7, 0.01)

# Mean and standard deviation pairs
params = [(0, 1), (0, 2), (3, 1)]

#plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization

for mu, sigma in params:
    y = normal(x, mu, sigma)
    plt.plot(x, y, label=f"μ={mu}, σ={sigma}")

plt.legend()  # Show the legend
plt.grid(True)  # Add a grid for better readability
plt.show()
```

```python
x = torch.linspace(0, 100, 101)

def loss_fn(b):
    return torch.sum((x - b) ** 2)

# 初始化 b
b = torch.tensor(20.0, requires_grad=True)

# 使用优化器
optimizer = torch.optim.SGD([b], lr=0.001)

# 迭代优化
for i in range(100):
    optimizer.zero_grad()
    loss = loss_fn(b)
    loss.backward()
    optimizer.step()

print(f"Optimized b: {b.item()}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter lambda (λ)
lam = 3

# x-axis range
x = np.arange(0, 15)

# Calculate Poisson probabilities
y = poisson.pmf(x, lam)

print(y)

# Plot the probabilities
plt.figure(figsize=(8, 6))

# The fix:  Use plt.vlines for older Matplotlib versions
plt.vlines(x, [0], y, color='blue', linewidth=2)  # Vertical lines

# Add markers (optional, but makes it look more like a stem plot)
plt.scatter(x, y, color='blue', s=50)  # Markers at the top of the lines

plt.title(f'Poisson Distribution (λ = {lam})')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability P(X=k)')
plt.xticks(x)
plt.grid(True)
plt.show()
```

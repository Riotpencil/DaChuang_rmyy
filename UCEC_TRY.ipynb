```python
from sklearn.cluster import KMeans
import numpy as np

def consensus_clustering(data, max_k=5, n_iterations=100, sample_fraction=0.8):
    """
    实现共识聚类
    Args:
        data (pd.DataFrame): 输入基因表达数据，行是基因，列是样品
        max_k (int): 最大聚类数 K
        n_iterations (int): 重复子采样和聚类的次数
        sample_fraction (float): 每次采样的样品比例

    Returns:
        dict: 包括共识矩阵、最佳 K 值和最终分组结果
    """
    n_samples = data.shape[1]
    consensus_matrices = {k: np.zeros((n_samples, n_samples)) for k in range(2, max_k + 1)}
    
    # 对每个 K 值执行多次采样和聚类
    for k in range(2, max_k + 1):
        for iter in range(n_iterations):
            # 随机采样样品
            sampled_indices = np.random.choice(n_samples, int(sample_fraction * n_samples), replace=False)
            sampled_data = data.iloc[:, sampled_indices]

            # 执行 KMeans 聚类
            kmeans = KMeans(n_clusters=k, random_state=None)
            labels = kmeans.fit_predict(sampled_data.T)  # 转置后每一行为一个样品
            
            # 更新共识矩阵
            for i, idx_i in enumerate(sampled_indices):
                for j, idx_j in enumerate(sampled_indices):
                    if labels[i] == labels[j]:
                        consensus_matrices[k][idx_i, idx_j] += 1
                        
        # 归一化共识矩阵
        consensus_matrices[k] /= n_iterations

    # 选择最佳 K 值
    avg_consensus = {k: np.mean(consensus_matrices[k]) for k in consensus_matrices}
    best_k = max(avg_consensus, key=avg_consensus.get)

    # 使用最佳 K 值对原始数据进行 KMeans 聚类
    final_kmeans = KMeans(n_clusters=best_k, random_state=None)
    final_labels = final_kmeans.fit_predict(data.T)  # 转置后每一行为一个样品

    return {
        "consensus_matrices": consensus_matrices,
        "best_k": best_k,
        "final_labels": final_labels  # 返回最终分组结果
    }
```

```python
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
from scipy.stats import kruskal
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def perform_survival_analysis_with_kruskal(labels, survival_data):
    """
    生存分析，包含 Kaplan-Meier 曲线绘制和 Kruskal-Wallis 检验
    Args:
        labels (np.ndarray): 样本的聚类标签
        survival_data (pd.DataFrame): 包含生存时间和状态的数据，需包含 'OS.time' 和 'OS' 列
    """
    kmf = KaplanMeierFitter()
    plt.figure(figsize=(8, 6))
    survival_times_by_cluster = []

    for cluster in sorted(np.unique(labels)):
        # 获取每个分组的样本
        cluster_mask = labels == cluster
        cluster_survival = survival_data[cluster_mask]
        
        # 绘制 Kaplan-Meier 生存曲线
        kmf.fit(cluster_survival["OS.time"], event_observed=cluster_survival["OS"], label=f"Cluster {cluster}")
        kmf.plot_survival_function()

        # 收集生存时间（用于 Kruskal-Wallis 检验）
        survival_times_by_cluster.append(cluster_survival["OS.time"].values)

    plt.title("Kaplan-Meier Survival Curves")
    plt.xlabel("Time")
    plt.ylabel("Survival Probability")
    plt.legend()
    plt.show()

    # 执行 Kruskal-Wallis 检验
    kruskal_result = kruskal(*survival_times_by_cluster)
    stat = kruskal_result.statistic  # 统计量
    p_value = kruskal_result.pvalue  # p 值

    print(f"Kruskal-Wallis Test Statistic: {stat}")
    print(f"P-value: {p_value}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_boxplot(filtered_mmp_data, final_labels, mmp_genes):
    """
    绘制两个组根据表达量数据的箱线图
    Args:
        filtered_mmp_data (pd.DataFrame): 基因表达数据，行是基因，列是样品
        final_labels (np.ndarray): 聚类标签，每个样品的聚类结果
        mmp_genes (list): 需要绘制箱线图的 MMP 基因列表
    """
    # 过滤出指定的 MMP 基因数据
    mmp_data = filtered_mmp_data.loc[mmp_genes, :]
    
    # 将数据转换为长格式：每一行是一个样品-基因组合
    mmp_data_long = mmp_data.T  # 转置，样品作为行，基因作为列
    mmp_data_long['Cluster'] = final_labels  # 添加聚类标签列

    # 使用 reset_index() 将样品的索引转换为列
    mmp_data_long = mmp_data_long.reset_index()
    mmp_data_long = mmp_data_long.melt(id_vars=["Cluster", "index"], value_vars=mmp_genes,
                                       var_name="Gene", value_name="Expression")
    mmp_data_long.rename(columns={"index": "Sample"}, inplace=True)

    # 绘制箱线图
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=mmp_data_long, x='Gene', y='Expression', hue='Cluster')
    plt.title("Boxplot of MMP Gene Expression by Cluster")
    plt.xlabel("MMP Genes")
    plt.ylabel("Expression Level")
    plt.xticks(rotation=45)
    plt.legend(title="Cluster", loc='upper left')
    plt.tight_layout()
    plt.show()

```

```python
def plot_boxplot(filtered_mmp_data, final_labels, mmp_genes):
    """
    绘制两个组根据表达量数据的箱线图，并计算每组的p值显示在图上。
    Args:
        filtered_mmp_data (pd.DataFrame): 基因表达数据，行是基因，列是样品
        final_labels (np.ndarray): 聚类标签，每个样品的聚类结果
        mmp_genes (list): 需要绘制箱线图的 MMP 基因列表
    """
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    from scipy.stats import ttest_ind

    # 过滤出指定的 MMP 基因数据
    mmp_data = filtered_mmp_data.loc[mmp_genes, :]
    
    # 将数据转换为长格式：每一行是一个样品-基因组合
    mmp_data_long = mmp_data.T  # 转置，样品作为行，基因作为列
    mmp_data_long['Cluster'] = final_labels  # 添加聚类标签列

    # 使用 reset_index() 将样品的索引转换为列
    mmp_data_long = mmp_data_long.reset_index()
    mmp_data_long = mmp_data_long.melt(id_vars=["Cluster", "index"], value_vars=mmp_genes,
                                       var_name="Gene", value_name="Expression")
    mmp_data_long.rename(columns={"index": "Sample"}, inplace=True)

    # 绘制箱线图
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=mmp_data_long, x='Gene', y='Expression', hue='Cluster')
    
    # 计算每个基因的p值
    for gene in mmp_genes:
        gene_data = mmp_data_long[mmp_data_long['Gene'] == gene]
        cluster_0 = gene_data[gene_data['Cluster'] == 0]['Expression']
        cluster_1 = gene_data[gene_data['Cluster'] == 1]['Expression']
        t_stat, p_value = ttest_ind(cluster_0, cluster_1, equal_var=False)

        # 在图上标注p值
        x_pos = mmp_genes.index(gene)
        y_max = gene_data['Expression'].max()
        plt.text(x_pos, y_max + 0.1, f"p={p_value:.3e}", ha='center', color='red', fontsize=10)

    plt.title("Boxplot of MMP Gene Expression by Cluster")
    plt.xlabel("MMP Genes")
    plt.ylabel("Expression Level")
    plt.xticks(rotation=45)
    plt.legend(title="Cluster", loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np

gene_data = pd.read_csv('exp.csv', index_col=0, sep=';') 
duplicates = gene_data.index[gene_data.index.duplicated()].unique()
df = gene_data.loc[~gene_data.index.duplicated(keep='first')]

genes = ["MMP3", "MMP10", "MMP9", "MMP11", "MMP12",
         "MMP13", "MMP15", "MMP19", "MMP23B", "MMP24", "MMP28"]
# Extract the expression data for the specified genes
extracted_data = gene_data.loc[genes]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
mmp_data_normalized = scaler.fit_transform(extracted_data.T)
mmp_data_normalized = pd.DataFrame(mmp_data_normalized.T,
                                   index=extracted_data.index,
                                   columns=extracted_data.columns)
clincal = pd.read_csv('clincal.csv', index_col=0, sep=';')
filtered_mmp_data = mmp_data_normalized.loc[:, mmp_data_normalized.columns.isin(clincal.index)]

from sklearn.cluster import KMeans
def consensus_clustering(data, max_k=5, n_iterations=100, sample_fraction=0.8):
    n_samples = data.shape[1]
    consensus_matrices = {k: np.zeros((n_samples, n_samples)) for k in range(2, max_k + 1)}
    for k in range(2, max_k + 1):
        for iter in range(n_iterations):
            # 随机采样样品
            sampled_indices = np.random.choice(n_samples, int(sample_fraction * n_samples), replace=False)
            sampled_data = data.iloc[:, sampled_indices]

            # 执行 KMeans 聚类
            kmeans = KMeans(n_clusters=k, random_state=None)
            labels = kmeans.fit_predict(sampled_data.T)  # 转置后每一行为一个样品

            # 更新共识矩阵
            for i, idx_i in enumerate(sampled_indices):
                for j, idx_j in enumerate(sampled_indices):
                    if labels[i] == labels[j]:
                        consensus_matrices[k][idx_i, idx_j] += 1

        # 归一化共识矩阵
        consensus_matrices[k] /= n_iterations

    # 选择最佳 K 值
    avg_consensus = {k: np.mean(consensus_matrices[k]) for k in consensus_matrices}
    best_k = max(avg_consensus, key=avg_consensus.get)

    # 使用最佳 K 值对原始数据进行 KMeans 聚类
    final_kmeans = KMeans(n_clusters=best_k, random_state=None)
    final_labels = final_kmeans.fit_predict(data.T)  # 转置后每一行为一个样品

    return {
        "consensus_matrices": consensus_matrices,
        "best_k": best_k,
        "final_labels": final_labels  # 返回最终分组结果
    }

results = consensus_clustering(filtered_mmp_data, max_k=5, n_iterations=1000)
final_labels = results["final_labels"]
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="seaborn")
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(2, 2, figsize=(16, 16))  # 创建 2x2 子图布局
clustermap_figures = []

# 生成带有树状图的热图并保存为图像
for idx, (k, matrix) in enumerate(results['consensus_matrices'].items()):
    clustermap = sns.clustermap(matrix, cmap="viridis", metric="euclidean", method="average")
    clustermap_figures.append((clustermap, k))  # 保存 clustermap 和 k 值
    plt.close(clustermap.fig)  # 关闭 clustermap 的原始图以避免重复显示

# 将 clustermap 图像拼接到一个图中
for idx, (clustermap, k) in enumerate(clustermap_figures):
    ax = axes[idx // 2, idx % 2]  # 确定子图位置
    ax.imshow(clustermap.fig.canvas.buffer_rgba())  # 添加图像
    ax.axis('off')  # 隐藏坐标轴
    ax.set_title(f"Consensus Matrix for k={k}", fontsize=14)  # 在子图上添加标题

plt.tight_layout()
plt.show()
```
